```python
import pandas as pd
df=pd.read_csv('/content/sample_data/california_housing_test.csv')
print(df.head(10))
```

```python
print(df.tail())
```

```python
print(df.info())
```

```python
nuevo_df=df.dropna()
print(nuevo_df.info())
```

```python
promedio=nuevo_df["total_bedrooms"].mean
print(promedio)
```

```python
for x in nuevo_df.index:
  if nuevo_df.loc[x,"total_bedrooms"]>900:
    nuevo_df.loc[x,"total_bedrooms"]=900
```

```python
import matplotlib.pyplot as plt
nuevo_df.plot()
plt.show()
```
